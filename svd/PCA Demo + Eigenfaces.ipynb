# Toy Example

```python
import numpy as np

Y = [[8,4,9,1,5],
[9,6,9,5,5],
[10,8,9,7,5],
[10,5,9,6,5],
[5,4,9,2,5],
[2,7,9,9,5],
[7,5,9,8,6],
[3,4,9,8,5],
[4,2,9,7,5],
[1,4,9,10,5]]
columnnames = ['Location','City','Society','Ambience','Vicinity to Airport']
rownames = ['House 1','House 2','House 3','House 4','House 5','House 6','House 7','House 8','House 9','House 10'] 
```

```python
import pandas as pd
Y = pd.DataFrame(Y, columns = columnnames, index=rownames)
Y
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=5) 
results = pca.fit(Y) 
```

```python
np.round(pca.explained_variance_,2)
```

```python
0.9*np.round(np.sum(pca.explained_variance_),2)
```

```python
np.round(pca.components_,2)
```

```python
np.round(results.components_[0]@results.components_[0],10)
```

```python
results.components_[0]@results.components_[1]
```

```python
np.round(results.components_[1]@results.components_[2],10)
```

```python
np.round(results.components_,decimals=3)
```

```python
Y_new = results.transform(Y)
```

```python
Y_new
```

```python
Y_new[:,0]@Y_new[:,1]
```

```python
np.corrcoef(Y_new[:,0],Y_new[:,1])
```

```python
np.dot(Y,results.components_.T)
```

```python
Y_new
```

```python
#Check the new features
np.round(Y_new,decimals=3)
```

```python
original = results.inverse_transform(Y_new)
np.round(original,2)
#see that the column with no variance has been returned as it is
```

# Eigenfaces

```python
#Loading LFW dataset from SKlearn
from sklearn.datasets import fetch_lfw_people
faces = fetch_lfw_people(min_faces_per_person=10)
print(faces.target_names)
print(faces.images.shape)
```

```python
#Importing standard libraries
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
```

```python
#Displaying a few images with labels
startindex = 0
fig, axi = plt.subplots(3, 3)
for i, axi in enumerate(axi.flat):
    axi.imshow(faces.images[i+startindex], cmap='bone')
    axi.set(xticks=[], yticks=[],
            xlabel=faces.target_names[faces.target[i+startindex]])
```

```python
faces.data.shape
```

```python
#Import PCA libraries and create a PCA instance
from sklearn.decomposition import PCA
pca = PCA(n_components = 1000)
pca.fit(faces.data)
```

```python
pca.components_[100]
```

```python
#Plotiing reduced dimensional data:
fig, axes = plt.subplots(3, 4, figsize=(9, 4),
                         subplot_kw={'xticks':[], 'yticks':[]},
                         gridspec_kw=dict(hspace=0.1, wspace=0.1))
for i, ax in enumerate(axes.flat):
    ax.imshow(pca.components_[i].reshape(62, 47), cmap='bone')
```

```python
#Plotiing reduced dimensional data:
fig, axes = plt.subplots(3, 4, figsize=(9, 4),
                         subplot_kw={'xticks':[], 'yticks':[]},
                         gridspec_kw=dict(hspace=0.1, wspace=0.1))
for i, ax in enumerate(axes.flat):
    ax.imshow(pca.components_[i+800].reshape(62, 47), cmap='bone')
```

```python
#commultative variance:
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
```

```python
# Compute the components and projected faces
pca = PCA(n_components = 200).fit(faces.data)
components = pca.transform(faces.data)
projected = pca.inverse_transform(components)

#COmparing the original and reconstructed faces:
# Plot the results
fig, ax = plt.subplots(2, 10, figsize=(10, 2.5),
                       subplot_kw={'xticks':[], 'yticks':[]},
                       gridspec_kw=dict(hspace=0.1, wspace=0.1))
#Change i below to see the results for other images
for i in range(10):
    ax[0, i].imshow(faces.data[i+50].reshape(62, 47), cmap='binary_r')
    ax[1, i].imshow(projected[i+50].reshape(62, 47), cmap='binary_r')
    
ax[0, 0].set_ylabel('full-dim\ninput')
ax[1, 0].set_ylabel('200-dim\nreconstruction')
```

#PCA VISUALIZATION: Another Example


```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
```

```python
rng = np.random.RandomState(1)
X = np.dot(rng.rand(2, 2), rng.randn(2, 200)).T
plt.scatter(X[:, 0], X[:, 1])
plt.axis('equal');
print(rng)
```

```python
C=np.cov(X.T)
print(C)
```

```python
evec, eval=np.linalg.eig(C)
print(eval)
print(evec)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(X)
```

```python
print(pca.components_)
print(pca.explained_variance_)
```

```python
#Dimension Reduction:
pca = PCA(n_components=1)
pca.fit(X)
X_pca = pca.transform(X)
print("original shape:   ", X.shape)
print("transformed shape:", X_pca.shape)
```
